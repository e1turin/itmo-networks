```python
txt = "ТИН" # Тюрин Иван Николаевич
codes = [0xD2, 0xC8, 0xCD] #encoded
str_msg = "".join(map(lambda n: f'{n:b}', codes))
msg = [int(c) for c in str_msg] #binary view

print(str_msg)
```

```python
С = 100 #Mbit/s | 1 Gbit/s
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Заданный двоичный сигнал
binary_message = str_msg

# Параметры кодирования
bit_duration = 1  # Длительность одного бита
carrier_frequency = 10  # Частота несущей

# Создание времени для временной диаграммы
time = np.arange(0, len(binary_message) * bit_duration, 0.01)

# Создание сигнала, модулированного АМ
am_signal = np.zeros(len(time))
for i, bit in enumerate(binary_message):
    if bit == '1':
        am_signal[i*100:(i+1)*100] = 1

# Построение временной диаграммы
plt.step(time, am_signal, where='post')
plt.xlabel('Время')
plt.ylabel('Амплитуда')
plt.title('Временная диаграмма кодирования (АМ)')
plt.grid(True)
plt.show()
```
